```python
# Load ML Pkgs
# Estimators
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

# Transformers
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load Dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/emotion_dataset_2.csv")
```

```python
# Data Cleaning
dir(nfx)
```

```python
# User handles
df['Clean_Text'] = df['Text'].apply(nfx.remove_userhandles)
```

```python
# Stopwords
df['Clean_Text'] = df['Clean_Text'].apply(nfx.remove_stopwords)
```

```python
df
```

```python
# Features & Labels
Xfeatures = df['Clean_Text']
ylabels = df['Emotion']
```

```python
#  Split Data
x_train,x_test,y_train,y_test = train_test_split(Xfeatures,ylabels,test_size=0.2,random_state=42)
```

```python
# Build Pipeline
from sklearn.pipeline import Pipeline
```

```python
# LogisticRegression Pipeline
pipe_lr = Pipeline(steps=[('cv',CountVectorizer()),('lr',LogisticRegression())])
```

```python
# Train and Fit Data
pipe_lr.fit(x_train,y_train)
```

```python
pipe_lr
```

```python
# Check Accuracy
pipe_lr.score(x_test,y_test)
```

```python
y_pred_for_lr=pipe_lr.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_for_lr))
```

```python
confusion_matrix(y_test,y_pred_for_lr)
```
